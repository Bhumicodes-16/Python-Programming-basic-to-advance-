```python
##Mthod overridding

class Animal:
    def speak(self):
        return "Sound of animal!"
    
class Dog(Animal):
    def speak(self):
        return "Woof!"
    
class Cat(Animal):
    def speak(self):
        return "Meow!"
    
def animal_speak(animal):
    print(animal.speak())
    

dog=Dog()
cat=Cat()

print(dog.speak())
print(cat.speak())
(animal_speak(dog))
    
    
```

```python
#Abstract class

from abc import ABC, abstractmethod

#Define an abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"    
```
