```python


## Encapsulation with Getter and Setter
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    ## getter method for name
    def get_name(self):
        return self.__name

    ## setter method
    def set_name(self, name):
        self.__name = name

    def get_age(self):
        return self.__age

    def set_age(self, age):
        self.__age = age


person = Person("Krish", 35)

print(person.get_name())
print(person.get_age())

person.set_age(35)


            
```
