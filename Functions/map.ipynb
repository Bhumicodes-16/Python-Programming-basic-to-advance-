```python
def square(x):
    return x*x

square(10)
```

```python
numbers=[1,2,3,4,5,6,7,8]

list(map(square,numbers))
```

```python
## Lambda function with map
numbers=[1,2,3,4,5,6,7,8]

list(map(lambda x:x*x,numbers))
```

```python
##map multiple iterables

numbers1=[1,2,3]
numbers2=[4,5,6]

added_number=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_number)
```

```python
## map( ) to convert list of string to integers

str_numbers=['1','2','3','4','5']
int_numbers=list(map(int,str_numbers))

print(int_numbers)
```

```python
words=["apple","cherry","banana"]
upper_word=list(map(str.upper,words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people=[
    {'name':'krish','age':32},
    {'name':'Jack','age':33}
]

list(map(get_name,people))
```
