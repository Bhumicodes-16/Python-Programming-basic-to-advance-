```python
## Arithmetic Operations

a=10
b=15

add_result=a+b
subtract_result=a-b
multiply_result=a*b
divide_result=a/b
floor_divide_result=a//b
modulus_result=a%b
exponent_result=a**b

print(add_result)  # Output: 25
print(subtract_result)  # Output: -5
print(multiply_result)  # Output: 150
print(divide_result)  # Output: 0.6666666666666666
print(floor_divide_result)  # Output: 0
print(modulus_result)  # Output: 10
print(exponent_result)  # Output: 1000000000000000
```
