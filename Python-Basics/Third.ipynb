```python
## Integer Example
age = 25
print(age)
print(type(age))
```

```python
## floating point datatype
height = 5.9
print(height)
print(type(height))
```

```python
## string datatype
name = "John Doe"
print(name)
print(type(name))
```

```python
## Boolean datatype
is_student = True
print(is_student)
print(type(is_student))
```

```python
is_true =bool(1)  # This will convert the integer 1 to a boolean, resulting in True
print(is_true)
print(type(is_true))
```

```python
a=10
b=10

a==b
type(a==b)  # Output: <class 'bool'>
```
