```python
## Understanding Variable Types
## Python is dynamically typed language, which means you don't have to declare the type of variable while declaring it. The interpreter infers the type of variable based on the value assigned to it.

age=25   #integer
height=5.9  #float
name="Nia"  #string
is_student=True  #boolean

print(type(age))  # <class 'int'>
print(type(height))  # <class 'float'>  
print(type(name))  # <class 'str'>
print(type(is_student))  # <class 'bool'>
```

```python
## Dynamic Typing
## Python allows the type of a variable to change during the execution of a program. This is known as dynamic typing.

var=10
print(var)  # Output: 10
print (type(var))  # Output: <class 'int'>

var="Hello"

print(var)  # Output: Hello
print (type(var))  # Output: <class 'str'>

var=3.14
print(var)  # Output: 3.14
print (type(var))  # Output: <class 'float'>

```
