```python
##Loop Control Statements

## break

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
##continue

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
##pass

## it is null operation which does noting

for i in range(5):
    if i==3:
        print("the num here is",i)
        pass
    print(i)
```

```python
##Nested loops

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```
